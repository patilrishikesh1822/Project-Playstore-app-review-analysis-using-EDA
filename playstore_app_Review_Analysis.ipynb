### Play Store App Review Analysis Using EDA Project Summary

The Play Store App Review Analysis project focuses on extracting meaningful insights from user reviews through **Exploratory Data Analysis (EDA)**. With millions of users sharing feedback daily, app reviews serve as a valuable source of information for understanding user satisfaction, identifying issues, and improving overall app performance. This project leverages data preprocessing, visualization, and statistical analysis to uncover patterns in user ratings and textual feedback.

The dataset used in this project typically consists of attributes such as app name, review text, user ratings (ranging from 1 to 5), review timestamps, and sometimes additional metadata like app category or version. The first step involves **data cleaning and preprocessing**, where missing values, duplicate entries, and irrelevant data are removed. Textual data is normalized by converting it to lowercase, removing punctuation, stopwords, and performing tokenization or lemmatization. These steps ensure consistency and prepare the data for meaningful analysis.

During the EDA phase, the **distribution of ratings** is examined to understand overall user sentiment. Most Play Store datasets show a skew toward higher ratings, with a majority of reviews falling in the 4- and 5-star categories. This indicates a general positivity bias, where satisfied users are more likely to leave feedback. However, lower ratings (1- and 2-star reviews) are particularly valuable as they often highlight critical issues such as app crashes, bugs, slow performance, or excessive advertisements.

To gain deeper insights, **sentiment analysis** is performed by mapping ratings to sentiment categories (positive, negative, and neutral). This helps quantify user opinions and compare them with textual feedback. In some cases, discrepancies are observed where the rating does not align with the sentiment expressed in the review text, indicating inconsistencies in user behavior.

Another key component of the analysis is **word frequency and text visualization**. Techniques such as word clouds and n-gram analysis are used to identify commonly occurring words and phrases. Positive reviews frequently include words like “good,” “easy,” “best,” and “useful,” reflecting satisfaction with usability and functionality. On the other hand, negative reviews often contain terms like “bug,” “crash,” “slow,” “ads,” and “update,” pointing to recurring technical and usability issues. These insights help developers prioritize problem areas.

The project also explores **category-wise analysis**, where apps are grouped based on their type, such as games, productivity, or social media. This reveals that gaming apps tend to receive a high volume of reviews but also more complaints related to ads and in-app purchases. Productivity apps are evaluated more on reliability and performance, while social apps often receive feedback related to updates and interface changes.

**Temporal analysis** is conducted to study trends over time. By analyzing review dates, the project identifies patterns such as spikes in negative reviews following app updates, which may indicate the release of buggy features. This helps in assessing the impact of updates on user satisfaction and guides better release management strategies.

Additionally, the analysis of **review length** shows that negative reviews are generally longer and more descriptive, providing detailed insights into user issues. In contrast, positive reviews are often short and less informative. Correlation analysis further highlights relationships between variables, such as a strong link between ratings and sentiment polarity.

In conclusion, this EDA-based project provides a comprehensive understanding of user feedback from Play Store reviews. It highlights key trends, common issues, and user preferences, enabling developers and stakeholders to make data-driven decisions. While EDA does not involve predictive modeling, it lays a strong foundation for advanced techniques like machine learning, sentiment classification, and recommendation systems, ultimately contributing to improved app quality and user experience.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Remove duplicates
apps.drop_duplicates(inplace=True)

# Clean Installs
apps['Installs'] = apps['Installs'].str.replace('[+,]', '', regex=True)
apps['Installs'] = pd.to_numeric(apps['Installs'], errors='coerce')

# Clean Price
apps['Price'] = apps['Price'].str.replace('$', '')
apps['Price'] = pd.to_numeric(apps['Price'], errors='coerce')

# Convert Reviews
apps['Reviews'] = pd.to_numeric(apps['Reviews'], errors='coerce')

# Drop nulls
apps.dropna(inplace=True)

# Merge datasets
merged = pd.merge(apps, reviews, on='App', how='inner')
````
